```python
df['CRIM'].fillna(df['CRIM'].mean(), inplace = True)
df['ZN'].fillna(df['ZN'].mean(), inplace = True)
df['INDUS'].fillna(df['INDUS'].mean(), inplace = True)
df['CHAS'].fillna(df['CHAS'].mean(), inplace = True)
df['AGE'].fillna(df['AGE'].mean(), inplace = True)
df['LSTAT'].fillna(df['LSTAT'].mean(), inplace = True)
```

```python
df.isnull().sum()
```

```python
sns.distplot(df['MEDV'])
```

```python
correlation = df.corr()
sns.heatmap(correlation, annot=True)
```

```python
sns.scatterplot(df['LSTAT'], df['MEDV'])
```

```python
sns.scatterplot(df['RM'], df['MEDV'])
```

```python
df.sample()
```

```python
X = np.c_[np.array(df['RM']), np.array(df['LSTAT'])]
```

```python
from sklearn.model_selection import train_test_split
```

```python
Y= np.array(df["MEDV"])
Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2 )
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
algo = LinearRegression()
```

```python
algo.fit(x_train,y_train)
```

```python
algo.intercept_
```

```python
algo.coef_
```

```python
y_predict=algo.predict(x_test)
```

```python
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
print("Testing Data Score")
rms_linear_test = np.sqrt(mean_squared_error(y_test, y_predict))
r2_score_test = r2_score(y_test, y_predict)
print("RMS SCORE", rms_linear_test)
print("R2", r2_score_test)
```

```python
y_predict_train= algo.predict(x_train)
```

```python
print("Train Data Score")
rms_linear_test = np.sqrt(mean_squared_error(y_train, y_predict_train))
r2_score_test = r2_score(y_train, y_predict_train)
print("RMS SCORE", rms_linear_test)
print("R2", r2_score_test)
```
